## What is NumPy?

NumPy (short for *Numerical Python*) is a scientific computing and mathematical library that we will be using to work with arrays and matricies of data. NumPy is very popular among machine learning practitioners and data scientists because fundamentally, all data can be thought of as an array of numbers and the NumPy package is the tailor-made for working with such structures. 

To start woking with NumPy we have to import the `numpy` library first. 

```
# `np` here is an alias for `numpy`
import numpy as np
```

```
np.__version__
```

Converting a list to a numpy array is easy:

```
my_list = [1, 2, 3, 4]

my_array = np.array(my_list)
```

```
type(my_list)
```

```
type(my_array)
```

In the resulting array, all elements belong to the same type (numeric or string):

```
my_array.dtype
```

```
mixed = [1, 2, 3, 'sring']
mixed_array = np.array(mixed)
```

```
print(mixed_array)
```

```
mixed_array.dtype
```

This is very a important property of NumPy array that distinguishes them from lists: all element of a NumPy array belong to the same type. This means that there is a very little memory overhead that we have to pay when creating such an array compared to lists -- it is sufficient to specify the array data type only once (in the case of a list you have to do it for every single element of the list). Also, NumPy arrays are represented by a continuous block of data in the computer memory while the elements of a list can be scattered all over the memory -- see Fig. 2-2 from the text. This makes it possible to do a lot of optimization with NumPy arrays, and this is why working with NumPy arrays is way more efficient than working with Python lists.

## Generating Random Arrays

### Integers

The NumPy module `random` allows you to create random data. For example, if you want to create a random integer number in the range from 0 to 9 (both values are inclusive), you can do it like this:

```
np.random.randint(10)
```

In this example, the integer values from 0 to 9 are said to be *uniformly distributed*. In simple language, it means that any one of these numbers has an equal chance to show up as an output of the `np.random.randint()` function. If we want to sound more scientific, we can also say that the function pulls random values from a uniform distribution of integers between 0 and 9.

```
for i in range(1000):
  print(f"{np.random.randint(10)} ", end='')
  if (i+1) % 50 == 0:
    print('\n')
```

(Note how we used the `'\n'` symbol to start a new line after every 50 printed values.) 

We can specify the smallest value that can appear in the output of the `np.random.randint()` function by providing two input values instead of one:

```
np.random.randint(1, 10)
```

Or, we can create a sequence of random numbers by providing the third argument:

```
np.random.randint(1, 10, 20)
```

We can even create a matrix of random integer values by providing the number of rows and columns as a tuple:

```
np.random.randint(1, 10, (5, 2))
```

You can read the full description of the `np.random.randint()` function in the [NumPy manual](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html) or use the built-in Python help system like this:

```
np.random.randint?
```

### Seed

If you generate a sequence of random values twice using exactly the same code the result is going to be different:

```
first = np.random.randint(0, 10, 10)
print(first)
```

```
second = np.random.randint(0, 10, 10)
print(second)
```

Sometimes this behavior is not desired because we want to be able to get a random but reproducible result. In this case one has to set the random seed before generating the random sequence.

```
np.random.seed(42)
third = np.random.randint(0, 10, 10)
print(third)
```

```
np.random.seed(42)
forth = np.random.randint(0, 10, 10)
print(forth)
```

Generally speaking, this is a very good idea to set the NumPy seed at the very beginning of your program, so that next time when you or somebody else run it all results are going to be the same.

## Shape

Every NumPy array has a certain number of dimensions. For example, a simple sequence of values has only one dimension:

```
one_dim = np.random.randint(1, 10, 10)
print(one_dim)
```

A matrix has two dimensions:

```
two_dim = np.random.randint(1, 10, (3, 7))
print(two_dim)
```

These dimensions are often called *axes* in Python. For example, in the case of the two-dimensional matrix the row-wise dimension is called *axis 0* and the column-wise dimension is called *axis 1*. 

Every axis of an array contains a certain number of elements. For example, in the case of the one dimensional array `one_dim`, there is only one axis and the number of elements is equal to 10, which is simply the length of the array. For the two-dimensional array `two_dim` the first axis contains 3 elements, which is simply the number of rows, and the second axis contains 7 elements, which is, of course, simply the number of columns. 

The combined information about the total number of axes and the number of elements along each axis of a NumPy array is called the *shape* of the array. The shape of any NumPy array can be determined by inspecting the `shape` attribute as follows:

```
print(one_dim.shape)
```

```
print(two_dim.shape)
```

Note that a NumPy array with the shape of `(10,)` is not the same as a `(1, 10)` array having the same entries. Let's reshape our `one_dim` array using the `reshape()` function: 

```
one_dim_reshaped = one_dim.reshape((1, 10))

print(one_dim)
print(one_dim_reshaped)
print(one_dim_reshaped.shape)
```

The same result can be achieved by adding a new axis explicitly:

```
one_dim_new_axis = one_dim[np.newaxis, :]
print(one_dim_new_axis)
print(one_dim_new_axis.shape)
```

Here are a few more examples on reshaping NumPy arrays:

```
one_dim.reshape((5, 2))
```

```
one_dim.reshape((2, 5))
```

In cases when one of the dimensions can be inferred unambiguously, you can replace it with `-1`. This tells Python: "Go and figure out this dimension on your own".

```
one_dim.reshape((-1, 5))
```

### View versus copy

An important thing to remember when working with NumPy arrays is that when you index or slice it you get back not a stand alone *copy* but just a *view* of the array. This view is still going to be a part of the array -- if you modify it, the original array will be modified as well.

```
array_slice = two_dim[:3, :2]
array_slice
```

```
array_slice[0, 0] = 0
array_slice
```

```
two_dim
```

This behavior is different from that of Python lists -- when you slice a list you get back a copy, not a view of the list:

```
some_list = [1, 5, 7, 9, 10]
```

```
list_slice = some_list[: 3]
list_slice
```

```
list_slice = [4, 6, 7]
```

```
some_list
```

If you really need to create a stand alone copy of an array, you should use the `copy()` function.

```
two_dim
```

```
array_slice = two_dim[:3, :2].copy()
array_slice
```

```
array_slice[0, 0] = 434223
array_slice
```

```
two_dim
```

As you can see, in this case changing the values of the slice does not result in changing the original array.

### Fancy indexing

The idea of *fancy indexing* is conceptually simple: it means passing an array of indices to access multiple elements at once. For example:

```
x = np.random.randint(1, 100, 10)
print(x)
```

Define an array of indices:

```
idx = np.array([2, 5, 4, 0, -1])
print(idx)
```

Select the specifying elements from the original array:

```
x[idx]
```

This indexing technique might be very useful. For example, let's say you want to shuffle two arrays of equal size in a consistent way.

```
y = np.random.randint(-99, 0, 10)
print(y)
```

Define an array of indices:

```
idx = np.arange(len(x))
print(idx)
```

Shuffle the indices:

```
np.random.shuffle(idx)
print(idx)
```

(Note that shuffling happens "in-place".) Now, index `x` and `y` using the array of shuffled indices:

```
x_sh = x[idx]
y_sh = y[idx]

print(x)
print(y)

print(x_sh)
print(y_sh)
```

## Joining NumPy Arrays

Joining of two or more NumPy arrays is primarily accomplished throught the routines `np.concatenate`, `np.vstack`, and `np.hstack`. 

### Concatenation

```
x = np.random.randint(1, 10, 5)
y = np.random.randint(-9, 0, 3)

print(x)
print(y)
```

```
x_and_y = np.concatenate((x, y))
print(x_and_y)
```

You can concatenate any number of arrays using `np.concatenate` function -- just pass a tuple with the names of all these arrays as an argument to the `np.concatenate` function. 

### Joining arrays by rows

To joing two arrays by rows just pass the tuple or list with the names of these arrays as an argument to the `np.vstack` function.

```
x = np.random.randint(1, 10, 10).reshape(2, 5)
y = np.random.randint(-9, 0, 10).reshape(2, 5)

print(x)
print(y)
```

```
join_by_rows = np.vstack((x, y))
print(join_by_rows)
```

### Joining arrays by columns

Similarly, to join two arrays in a column-wise fashion, use the `np.hstack` function:

```
join_by_cols = np.hstack((x, y))
print(join_by_cols)
```

## Computation on NumPy Arrays

Suppose you want to add two arrays of equal size:

```
a = np.random.randint(1, 10, 10)
b = np.random.randint(1, 10, 10)
```

If you are coming from Java or C++ background you might be tempted to do something like this:

```
def add_two(a, b):
  
  result = np.array([], dtype=int)

  for i in range(len(a)):
    sum = a[i]+b[i]
    result = np.append(result, sum)

  return result
```

```
add_two(a, b)
```

```
%timeit _ = add_two(a, b)
```

This is pretty slow. The slowdown happens because at each execusion of the `for` loop, Python needs to look up the data types of the array elements and to decide how to do the `+` operation for the given data type. These technical details are not that important. What is important is the following message: when working with NumPy arrays you must stay as far away as possible from using `for`-loops in this manner! Use *vectorize* operations instead. Here is an example:

```
a+b
```

Or, you can use a specialized NumPy function (the `+` is just a convenient wrapper used for this function):

```
np.add(a, b)
```

The addition is performed in the element-by-element fashion and the result of this operation is a new NumPy array that has exactly the same dimension as input arrays. 

Now, let's compare the speed.

```
def add_two_vect(a, b):
  result = a + b
  return result
```

The sums computed by the two functions are exactly the same:

```
s1 = add_two(a, b)
s2 = add_two_vect(a, b)

print(s1)
print(s2)
```

```
%timeit _ = add_two_vect(a, b)
```

This is a significant speed up! The vectorized version is about 100 times faster!

One can use similar approach to do subtraction, mutliplication, division, and many other operations.

```
print(a - b)
print(a/b)
print(a*b)
print(a**3)
print(3**a)

print(np.abs(a-b))
print(np.sin(b))
print(np.log(b))
print(np.exp(a))
```

These vectorized NumPy functions are known as the *Universal Functions* of the NumPy package, or *ufuncs* for short.

See Table 2-2 for the full list of arithmetic operations implemented in NumPy. Also, read about NumPy ufunc's for computing absolute values, trigonomertic functions, exponents and logarithms on pp. 54-55 of the textbook.

## NumPy Broadcasting Rules

Cosider a situation when you need to add a single number (also knows as a scalar) to every element of some NumPy array whose length is larger than one. How would you do that? 

```
x = np.arange(3)
n = 5 
print(x)
```

One thing that you might want to try is to turn the scalar `n` into a NumPy array whose dimesion matches that of `x`. In our example, it means that we need to stretch or duplicate the value `5` into the array `[5, 5, 5]`:

```
n_array = np.repeat(n, len(x))
print(n_array)
```

```
x + n_array
```

It is a lot of coding! Besides, it is really necessary to duplicate the number `5`? There must be a better way of doing it! Fortunately, NumPy allows you to just add the number to the array directly, like this:

```
x + n
```

The result is the same but it is much less coding! This magic is possible because of the so-called NumPy *broadcasting* which is a set of rules for applying binary ufuncs (addition, subtraction, division, etc.) on arrays of different sizes. In order to visulize and understand the broadcasting operation, you can imagine that in order to match the sizes of the input arrays, Python performs necessary duplication or stretching similar to the one that we did with the scalar value `5`. In reallity it does not happen because it would be an unnecessary waist of resources! But it is a good mental picture that can help us to understand the process.

Here are a few more examples:

```
ones3=np.ones((3, 3), dtype=np.uint8)
print(ones3)
```

What if we add `x` to this matrix? Recall that `x` looks like this:

```
print(x)
```

Adding these two yields:

```
ones3 + x
```

It worked! The row `x` has been added to every row of `ones3`. You can imagine that Python stretched the row `x` by creating 2 extra copies of it and stacking 3 versions of `x` vertically.  Here is the picture that helps us to visualize the process:

broadcasting.png

Let's try the last example from the diagram above:

```
x[: , np.newaxis] + x
```

As you can see, in this situation the broadcasting works as if both of the arrays have been streched in the ways shown in the diagram.

These examples are going to be more than enough for the purpose of our class, but if you want to know how exactly the broadcasting is done in the most general situation then here is the set of rules:

**Rules of Broadcasting:** 
1. If the two arrays differ in their number of dimensions, the shape of the one with fewer dimensions is *padded* with ones on its leading (left) side. 
2. If the shapes of the two arrays do not match in any dimension, the array with shape equal to 1 in that dimension is stretched to match the other shape.
3. If in any dimension the sizes disagree and neither is equal to 1, an error is raised.

```
a = np.ones((4, 3), dtype=np.uint8)
b = np.arange(3)

print(a)
print(b)
```

```
print(a.shape)
print(b.shape)
```

```
a + b
```

## Comparisons of NumPy Arrays

### Elementwise comparison

It is possible to do element-by-element comparisson of NumPy arrays.

```
np.random.seed(42343)

a = np.random.randint(1, 10, 10)
b = np.random.randint(1, 10, 10)

print(a)
print(b)
```

The result is a NumPy array whose values are equal to either `True` or `False`. 

```
c = a < b
print(c)
```

This type of array is called *boolean* NumPy array.

```
print(c.dtype)
```

```
print(a != b)
```

Broadcasting works in the usual way:

```
print(a > 3)
```

### Aggregating functions

There are a couple of very useful aggregation functions for working with boolean arrays:
* `np.any()` -- checks if any element of the the array is `True`;
* `np.all()` -- checks if all elements of the array are `True`;
* `np.count_nonzero()` or `np.sum()` -- counts the number of `True` elements.

```
print(a)
```

Is there any element in `a` that is larger than 3?

```
(a>3).any()
```

Are all elements of `a` larger than 3?

```
(a>3).all()
```

Is there any element of `a` larger than 10?

```
(a>10).any()
```

Are all elements of `a` larger than or equal to zero?

```
(a>=0).all()
```

This can be used within an `if`-statements:

```
if (a>=0).all():
  print("All element of the array are positive!")
else:
  print("There are negative elements in the array!")
```

What if you want to know how many elements in a given array meet a certain condition? For example, how many elements in array `a` are larger than 3? To answer this question we can use the `np.sum()` aggregation function on the boolean array: when this function is computed every `True` value is replaced with a one and every `False` value is replaced with a zero. 

```
print(a)
(a>3).sum()
```

Alternatively, we can use `np.count_nonzero()` function. It's about 3 times faster:

```
np.count_nonzero(a>3)
```

### Boolean logic on NumPy arrays

When combining two boolean arrays do not use the standard `and`, `or`, and `not` operators! Normally, you want to combine your arrays elementwise and for that you need the use the so-called *bitwise* version of the logical operators:

```
print(b) 
```

Let's find the elements of `b` that are larger than 3 but smaller than 8. We will use the `&` operators which represents a bitwise AND:

```
print((b > 3) & (b < 8))
```

Let's find the elements of `b` that are either larger than 7 or smaller than 4. We will use the `|` operators which represents a bitwise OR:

```
d = (b > 7) | (b < 4)
print(d)
```

To negate a given boolean array use the `~` operator representing a bitwise NOT:

```
print(~d)
```

### Boolean masks

A great feature of NumPy is that the boolean NumPy arrays can be used as boolean mask used to select the elements of an array meeting a certain logical condition. For example, to select from `b` all elements that are larger than 6 we can do this:

```
mask = b > 6

print(mask)
```

Now, we can index `b` using this mask:

```
b[mask]
```

Recall that `b` looks like this:

```
print(b)
```

In order for this procedure to work, the shapes of the original array and the mask must be the same. Otherwise you will get an error. 

Another example with a two-dimensional array:

```
np.random.seed(323134)

a2d = np.random.randint(1, 100, (4, 6))
print(a2d)
```

Let's select all values between 30 and 50:

```
mask2d = (a2d >= 30) & (a2d <= 50)

print(mask2d)
```

```
a2d[mask2d]
```

Note that the resulting array is one-dimensional.
